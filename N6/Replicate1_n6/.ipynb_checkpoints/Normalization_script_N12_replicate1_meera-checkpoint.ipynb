```python
# extra_time=0
# time_array= np.array([0, 140, 290, 420, 570, 755, 950, 1645])

# # Dilution_half_life_hrs= 6.739
# Dilution_half_life_hrs= 6.0
# Cfactor= 100



extra_time=0
time_array= np.array(([0,3*60,5*60+15,7*60+50+extra_time,11*60+extra_time,14*60+50+extra_time,21*60+30+extra_time,58*60+extra_time]))
Dilution_half_life_hrs= 12.0
Cfactor= 100

## Commonly used variables are already defined here 
def dataframe_with_n(n): 
    
      return  tmt5[tmt5['Nitrogens']==n]

## Commonly used functions later on 

def normalized_array_for_plotting(dilution_half_life_h, active_degradation_rate_per_min, nitrogen_number, K, extra_time): 
    #time_array= np.array(([0,3*60+45+extra_time,5*60+30+extra_time,7*60+30+extra_time,15*60+30+extra_time,20*60+30+extra_time,25*60+20+extra_time,43*60+20+extra_time]))
    D=np.log(2)/ dilution_half_life_h/60
    Dt= D+active_degradation_rate_per_min
    n=nitrogen_number
    term1= Dt
    term2= n*Cfactor*D*K
    constant= term1-term2
    final= (term1*np.exp(-term2*time_array)-term2*np.exp(-term1*time_array))/constant
    sum_final=final.sum()
    final1= final/sum_final
    return final1

```

```python
def vector_sum(x): 
    i=0 
    array_new= []
    while i<(len(x)-1): 
        array_new.append(np.sum(x[i:i+8])) 
        i=i+8
    #print array_new
    return np.repeat(array_new,8)

def dilution_fitting(X, a):

    t,n= X
    c= Cfactor
    vector1= (a*np.exp(-n*t*c*a)- n*c*a*np.exp(-a*t))
    vector2= vector_sum(vector1)
    vector3= vector1/vector2
    return vector3




membrane_proteins= tmt5[tmt5['membrane_confirmation_2']==True]
membrane_proteins.shape

# Fit a dilution ate to the membrane proteins 

df_temp= membrane_proteins[['0_med_ave', '1_med_ave', '2_med_ave','3_med_ave','4_med_ave','5_med_ave','6_med_ave','7_med_ave']]
values1= df_temp.values.flatten()
time_array1= np.tile(time_array, len(df_temp))
nitrogens= membrane_proteins['Nitrogens'].values
n_array= np.repeat(nitrogens,8)

array1_dilution,array2_dilution=curve_fit(dilution_fitting, (time_array1, n_array), values1, p0=[np.log(2)/Dilution_half_life_hrs/60])
dilution_rate_calculated= np.log(2)/array1_dilution[0]/60
print ( 'the fitted dilution rate to the membrane proteins is: %f' %(dilution_rate_calculated))
```

```python
from scipy.optimize import curve_fit
def vector_sum(x): 
    i=0 
    array_new= []
    while i<(len(x)-1): 
        array_new.append(np.sum(x[i:i+8])) 
        i=i+8
    #print array_new
    return np.repeat(array_new,8)

Dilution_rate= dilution_rate_calculated 
def degradation_new(X, a):
    kd= np.log(2)/ (Dilution_rate*60)
    t,n= X
    c= Cfactor
    vector1= (a*np.exp(-n*t*c*kd)- n*c*kd*np.exp(-a*t))
    vector2= vector_sum(vector1)
    vector3= vector1/vector2
    return vector3




grouped=tmt5.groupby(['Protein ID'])
protein_id=[]
entry=[]
# gene_name=[]
half_life=[]
degradation_rates_mins=[]
for name, group in grouped: 
#     print (name)
    df_temp=grouped.get_group(name)
    df_temp1= df_temp[['0_med_ave', '1_med_ave', '2_med_ave','3_med_ave','4_med_ave','5_med_ave','6_med_ave','7_med_ave']]
    # y and x data 
    values1= df_temp1.values.flatten()
    time_array1= np.tile(time_array, len(df_temp))
    nitrogens= df_temp['Nitrogens'].values
    n_array= np.repeat(nitrogens,8)
    
    protein_id.append(name)
    entry.append(df_temp.iloc[0]['Entry'])
#     gene_name.append(df_temp.iloc[0]['Gene names '])

    try: 
        array1,array2=curve_fit(degradation_new, (time_array1, n_array), values1, p0=[np.log(2)/dilution_rate_calculated/60])
        degradation_readjusted_value= array1[0]-array1_dilution[0]+ np.log(2)/(Dilution_half_life_hrs*60)
        degradation_rates_mins.append(degradation_readjusted_value)

        half_life.append(1*np.log(2)/degradation_readjusted_value/60)
    except:
        print (name)
        degradation_rates_mins.append(float('Nan'))
        half_life.append(float('Nan'))
```
